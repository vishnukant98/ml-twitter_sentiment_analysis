```python
def new_dataFrame(df):
    new_df = pd.DataFrame()
    new_df["text"] = df["text"]
    new_df["airline_sentiment"] = df["airline_sentiment"]
    return new_df
```

```python
Y_train = df["airline_sentiment"]
X_train = df["text"]
X_test = df1["text"]
airlines = df["airline"]
```

```python
airlines.head(5)
```

```python
airlines_set = set(airlines)
```

```python
from nltk.tokenize import word_tokenize
```

```python
documents = []
for i in range(len(X_train)):
    documents.append((word_tokenize(X_train[i]),Y_train[i]))
documents[0:5]
```

```python
documents_test = []
for i in range(len(X_test)):
    documents_test.append(word_tokenize(X_test[i]))
documents_test[0:5]
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith("J"):
        return wordnet.ADJ
    elif tag.startswith("V"):
        return wordnet.VERB
    elif tag.startswith("N"):
        return wordnet.NOUN
    elif tag.startswith("R"):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words("english"))
punctuations = list(string.punctuation)
stops.update(punctuations)
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
lemmatizer = WordNetLemmatizer()
```

```python
def clean_text(words):
    output_words = []
    for w in words:
        if w not in airlines_set:
            if w.lower() not in stops:
                pos = pos_tag(w)
                clean_words = lemmatizer.lemmatize(w,pos = get_simple_pos(pos[0][0]))
                output_words.append(clean_words)
    return output_words
```

```python
documents = [(clean_text(words),category) for words, category in documents]
documents_test = [(clean_text(words)) for words in documents_test]
```

```python
all_words = []
for doc in documents:
    all_words += doc[0]
```

```python
import nltk
freq = nltk.FreqDist(all_words)
freq
```

```python
common = freq.most_common(3000)
```

```python
features = [i[0] for i in common]
features
```

```python
def get_feature_dict(words):
    current_features = {}
    word_set = set(words)
    for w in features:
        current_features[w] = w in word_set
    return current_features
```

```python
training_documents = [(get_feature_dict(words), category) for words, category in documents]
```
